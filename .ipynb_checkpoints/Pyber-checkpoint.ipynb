```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
city_data = pd.read_csv('city_data.csv')
city_data.head()
```

```python
ride_data = pd.read_csv('ride_data.csv')
ride_data.head()
```

```python
merged_driver_data = pd.merge(ride_data, city_data, on='city', how='outer')
merged_driver_data.head()
```

```python
# Set Up for Bubble Chart
groupby_city = merged_driver_data.groupby('city')
fare_city = groupby_city[['fare']].mean()
avg_fare_city = fare_city.rename(columns={'fare' : 'average_fare'})
avg_fare_city.head()
```

```python
rides_city = groupby_city[['ride_id']].count()
rides_per_city = rides_city.rename(columns={'ride_id' : 'rides_per_city'})
rides_per_city.head()
```

```python
rides_avgfare_city = rides_per_city.join(avg_fare_city)
x_axis = rides_avgfare_city['rides_per_city']
y_axis = rides_avgfare_city['average_fare']
```
